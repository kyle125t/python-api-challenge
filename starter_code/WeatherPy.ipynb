```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "../output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)

# Set Date for plots below

date = time.strftime("%m/%d/%Y")
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []
countries = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=300)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=300)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    country = citipy.nearest_city(lat_lng[0], lat_lng[1]).country_code
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city.title())
        countries.append(country.upper())
    
# Print the city count to confirm sufficient count
print(len(cities))
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
base_url = f"http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID={weather_api_key}&q="
# Create dataframe based off examples provided
weather_check = pd.DataFrame({"City": cities,
                            "Lat": "",
                            "Lng": "",
                            "Max Temp": "",
                            "Humidity": "",
                            "Cloudiness": "",
                            "Wind Speed": "",
                            "Date": "",
                            "Country": countries})
# Help restrict searches by sets
record = 1
rec_set = 1

print("Beginning Data Retrieval")
print("------------------------------")
# Searching by row and index for respective information from URL query and adding it to new DF
for index, row in weather_check.iterrows():

    query_url = base_url + row["City"]
    response = requests.get(query_url).json()
    try:
        print(f"Processing Record {record} of Set {rec_set} | {response['name']}")
        weather_check.loc[index, "Lat"] = response["coord"]["lat"]
        weather_check.loc[index, "Lng"] = response["coord"]["lon"]
        weather_check.loc[index, "Cloudiness"] = response["clouds"]["all"]        
        weather_check.loc[index, "Max Temp"] = response["main"]["temp_max"]
        weather_check.loc[index, "Humidity"] = response["main"]["humidity"]
        weather_check.loc[index, "Wind Speed"] = response["wind"]["speed"]
        weather_check.loc[index, "Date"] = response["dt"]
        record += 1
    except (KeyError, IndexError):
        print("City not found... Continuing...")
    if record == 25:
        rec_set += 1
        record = 1
        # Prevent going over API limits with a sleep timer between requests
        time.sleep(1.1)

print("------------------------------")
print("Data Retrieval Complete")

# Convert Unix values in Date column to be more easily read
weather_check["Date"] = pd.to_datetime(weather_check["Date"], unit = "s")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# Removing duplicates from the list, setting inplace to lock changes moving forward
weather_check.drop_duplicates(inplace = True)
```

```python
# Removing blank values from Humidity (which was frequently giving me problems)
weather_check.drop(weather_check.loc[weather_check["Humidity"]== ''].index, inplace=True)
```

```python
weather_check.to_csv(output_data_file)
```

## Linear Regression

```python
# Set hemispheres

north_hemi = weather_check.loc[weather_check["Lat"] >= 0.0]
north_hemi.reset_index(inplace=True)

south_hemi = weather_check.loc[weather_check["Lat"] < 0.0]
south_hemi.reset_index(inplace=True)
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = north_hemi["Lat"].astype(int)
y_values = north_hemi["Max Temp"].astype(int)
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(north_hemi["Lat"], north_hemi["Max Temp"], c = "blue", alpha = 0.5, edgecolors = "black")
plt.title("Northern Hemisphere - Max Temp vs. Latitude Linear Regression")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.annotate(line_eq, (10, -30), fontsize = 15, color = "red")
plt.plot(x_values,regress_values,"r-")
print(f"The r-squared is: {rvalue**2}")
```
